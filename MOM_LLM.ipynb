# Loading Model and Downloading it

```python
tokenizer = AutoTokenizer.from_pretrained("google/gemma-2-9b", token=os.environ["HF_TOKEN"])
model = AutoModelForCausalLM.from_pretrained(
    "google/gemma-2-9b",
    device_map="auto",
    torch_dtype=torch.bfloat16,
    token=os.environ["HF_TOKEN"]
)
```

# Function for generating response

```python
def generate_response(prompt, max_new_tokens=5000):
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=5500).to(model.device)
    outputs = model.generate(**inputs, max_new_tokens=max_new_tokens, do_sample=True, temperature=0.2)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

# Functions for prompting and getting output

```python
def summarize_minutes(text):
    prompt = f"""Summarize the following meeting minutes:

    {text}

    Summary:"""
    return generate_response(prompt)

def extract_action_items(text):
    prompt = f"""Extract action items from the following meeting minutes:

    {text}

    Action items:
    1."""
    return generate_response(prompt)

def answer_question(text, question):
    prompt = f"""Given the following meeting minutes:

    {text}

    Answer the following question:
    Q: {question}
    A:"""
    return generate_response(prompt)

def categorize_action_items(action_items, categories):
    prompt = f"""Categorize the following action items into the given categories:

    Action items:
    {action_items}

    Categories:
    {', '.join(categories)}

    Categorized action items:"""
    return generate_response(prompt)
```

```python
def load_mom_documents(folder_path):
    mom_documents = {}
    for filename in os.listdir(folder_path):
        if filename.endswith(".docx"):
            file_path = os.path.join(folder_path, filename)
            content = read_docx(file_path)
            mom_documents[filename] = content
    return mom_documents
```

# Function for reading Docs Files

```python
def read_docx(file_path):
    doc = Document(file_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)
```

```python
def process_query(query, mom_documents):
    combined_content = "\n\n".join([f"Document: {filename}\n{content}" for filename, content in mom_documents.items()])
    prompt = f"""Given the following meeting minutes from multiple documents:

    {combined_content}

    User Query: {query}

    Response: """

    return generate_response(prompt, max_new_tokens=1024)
```

# Prompting and getting output

```python
def main():
    #folder_path = input("Enter the path to the folder containing MOM documents: ")
    mom_documents = load_mom_documents('./Docs/')

    if not mom_documents:
        print("No .docx files found in the specified folder.")
        return

    print(f"\nLoaded {len(mom_documents)} MOM documents.")
    print("You can now ask questions about the meetings, request summaries, or ask for specific information.")
    print("Type 'quit' to exit.")

    while True:
        query = input("\nWhat would you like to know about the meetings? ")

        if query.lower() == 'quit':
            break

        response = process_query(query, mom_documents)
        print("\nResponse:")
        print(response)

if __name__ == "__main__":
    main()
```
